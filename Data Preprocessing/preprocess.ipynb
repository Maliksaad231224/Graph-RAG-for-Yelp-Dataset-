```python
import pandas as pd
```

```python
customer = pd.read_csv(r'olist_customers_dataset_enhanced_sample.csv')
payments = pd.read_csv(r'olist_order_payments_dataset_enhanced_sample.csv')
reviews = pd.read_csv(r'olist_order_reviews_dataset_enhanced_sample.csv')
orders = pd.read_csv(r'olist_orders_dataset_enhanced_sample.csv')
products = pd.read_csv(r'olist_products_dataset_enhanced_sample.csv')
sellers = pd.read_csv(r'olist_sellers_dataset_enhanced_sample.csv')
translations = pd.read_csv(r'product_category_name_translation_sample.csv')
items = pd.read_csv(r'olist_order_items_dataset_enhanced_sample.csv')
```

```python
print("customer columns:",customer.columns)
print("payments columns:",payments.columns)
print("reviews columns:",reviews.columns)
print("orders columns:",orders.columns)
print("products columns:",products.columns)
print("sellers columns:",sellers.columns)
print("translations columns:",translations.columns)
print("items columns:",items.columns)
```

```python
customer=customer[:20000]
```

```python
valid_customers = customer["customer_id"].unique()
orders = orders[orders["customer_id"].isin(valid_customers)]

valid_orders = orders["order_id"].unique()
items = items[items["order_id"].isin(valid_orders)]

payments = payments[payments["order_id"].isin(valid_orders)]

reviews = reviews[reviews["order_id"].isin(valid_orders)]

valid_products = items["product_id"].unique()
products = products[products["product_id"].isin(valid_products)]

valid_sellers = items["seller_id"].unique()
sellers = sellers[sellers["seller_id"].isin(valid_sellers)]

if "product_category_name" in items.columns:
    valid_categories = items["product_category_name"].dropna().unique()
    translations = translations[translations["product_category_name"].isin(valid_categories)]

```

```python
# Clip datasets to 30,000 records (except customers)
MAX_RECORDS = 30000

if len(orders) > MAX_RECORDS:
    orders = orders[:MAX_RECORDS]
    print(f"Orders clipped to {MAX_RECORDS} records")

if len(items) > MAX_RECORDS:
    items = items[:MAX_RECORDS]
    print(f"Items clipped to {MAX_RECORDS} records")

if len(payments) > MAX_RECORDS:
    payments = payments[:MAX_RECORDS]
    print(f"Payments clipped to {MAX_RECORDS} records")

if len(reviews) > MAX_RECORDS:
    reviews = reviews[:MAX_RECORDS]
    print(f"Reviews clipped to {MAX_RECORDS} records")

if len(products) > MAX_RECORDS:
    products = products[:MAX_RECORDS]
    print(f"Products clipped to {MAX_RECORDS} records")

if len(sellers) > MAX_RECORDS:
    sellers = sellers[:MAX_RECORDS]
    print(f"Sellers clipped to {MAX_RECORDS} records")

if len(translations) > MAX_RECORDS:
    translations = translations[:MAX_RECORDS]
    print(f"Translations clipped to {MAX_RECORDS} records")

print("\nFinal record counts:")
print(f"Customers: {len(customer)}")
print(f"Orders: {len(orders)}")
print(f"Items: {len(items)}")
print(f"Payments: {len(payments)}")
print(f"Reviews: {len(reviews)}")
print(f"Products: {len(products)}")
print(f"Sellers: {len(sellers)}")
print(f"Translations: {len(translations)}")
```
